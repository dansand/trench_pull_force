# Mathematical background & analysis notes


As described in the manuscript, the vertically-integrated horizontal force balance can be written as:

\begin{equation}
\bar \sigma_{xx}(x_1) - \bar \sigma_{xx}(x_0) + \int_{x_0}^{x_1} \tau_{xz}(x, z_c) dx = 0 \;\;\mathrm{{\bf(1)}}
\end{equation}

Equation **(1)** represents horizontal static equilibrium on arbitrary 2D domain, where the boundaries are aligned with the coordinate system, as shown in Figure 1 of the manuscript. The terms in Equation **(1)** represent net horizontal forces per unit length. These force comonents derive from integrals of the horizonal compenent of tractions across the vertical and basal boundaries. Because the domain is aligned with the coordinate system, the horizontal component of the traction simplifies to a single stress component multiplied by either a 1 or a negative 1 depending on the outward normal. The upper boundary is considered shear stress free and makes no contribution to the horizontal force balance. Body forces do not appear directly in Equation **(1)** as gravity is assumed to be vertical everywhere. 

Equation **(1)** can be read as: the difference in the integrated horizontal traction due to basal shear stress ($\tau_{xz}$) from $x_0$ to $x_1$, at $z = z_c$, equals the difference in the vertically integrated horizontal traction due to the normal stress $\sigma_{xx}$. The **overbars** in Equation **(1)** represent an integration from an upper equipotential surface, such as sealevel (which is the level assumed in the manuscript) down to another equipotential surface which is referred to as the compensation level ($z_c$). This level is arbitrary in a sense, but the assumption is that at this level horizontal gradients in pressure (or vertical normal stress) are negligible (hydrostatic approximation).  This is the standard approximation in thin-plate flexural model. 

Equation **(1)** is still completely valid if there are pressure gradients along the basal boundary, but a significant pressure gradient would imply significant dynamic support of the topography, and would complicate the interpretation of the terms in Equation **(1)**. Note that in terms of trying to understand and isolate the magnitude of the trench pull force, what is relevant is whether the dynamic pressure contribution i.e., $\Delta P_{z_c}$ is significant relative to the trench pressure deficit $\Delta P_T  = (\rho_m - \rho_w) g w_T$, across the same implied length scale, given by the distance from the trench axis to an isostatic column in the trailing plate. 

----------------

We can rewrite $\sigma_{xx}$ as: 

$\sigma_{xx} = \tau_{xx} + \sigma_I= \tau_{xx} + \sigma_{zz} - \tau_{zz}$

Where $\sigma_I = -P$ is the mean stress. Stress tensor quantities are assumed negative in compression. Substituting in Equation **(1)** gives:

\begin{equation}
\Delta (\overline{\tau_{xx}-\tau_{zz}})  +  \Delta \bar\sigma_{zz}  +  \int_{x_0}^{x_1} \tau_{xz}(x, z_c) = 0 \;\;\mathrm{{\bf(2)}}
\end{equation}

Where the $\Delta$ symbol is a shorthand used to represent the difference between the vertical integrals at points $x_0$ and $x_1$. In Equation **(1)** & **(2)** a positive change in any of the quantities, from $x_0 \rightarrow x_1$ represents a net force to the right. For instance a positive change in $\Delta (\overline{\tau_{xx}-\tau_{zz}})$ represents a reduction in the integrated horizontal normal stress relative to the vertical, we can say that the in-plane stress is becoming less compressional in the positive $x$ direction.  

In the manuscript, these 3 terms are symbolised: 

\begin{equation}
\Delta F_{D}  -  \Delta \mathrm{GPE}^*  +  F_B = 0 \;\; \mathrm{{\bf(3)}}
\end{equation}

Where the definition imposed is: $\Delta \mathrm{GPE}^* = -\Delta \bar\sigma_{zz}$

The sign change implemented here makes the sign of $\Delta \mathrm{GPE}^* $ consistent with the definition for the "true" GPE (which is based on positive quantity  - pressure, rather than a negative quantity, the vertical normal stress).  This means that a positive change in $\Delta \mathrm{GPE}^*$ from  $x_0 \rightarrow x_1$, now represents a net force to the left. The reason for asterisk on the $\mathrm{GPE}^*$, is discussed in the paper. Fundamentally, $\mathrm{GPE}^*$ represents the (negative of) the vertical integral of "true" vertical normal stress $\bar\sigma_{zz}$. Meanwhile, the "true" $\mathrm{GPE}$ is the integral of the lithostatic pressure ($\bar P_L$). In general, $\mathrm{GPE}^* \neq \mathrm{GPE}$. The quantities are only equal when the vertical normal stress in both columns is lithostatic. This is not the case for trench topography!. The paper shows that one can interpret  $\bar\sigma_{zz}$ in as a GPE-like quantity, even for non-isostatic columns, if we use a corrected density ($\rho^*$) that accounts for horizontal gradients in the vertical shear stress: $\frac{\partial \tau_{zx}}{\partial x}$.

In the manuscript, the following scaling relationship is proposed for the trench pull force:  

\begin{equation}
\text{trench pull force} \equiv \Delta \mathrm{GPE}^* = (\rho_m - \rho_w) g w_T ( \frac{z'_m}{2}) \approx \Delta P_T z'_{np}
\end{equation}

Where $w_T$ is the deflection at the trench axis relative to the isostatic level, $\Delta P_T$ is referred to as the trench pressure deficit, $z'_m$ is the mechanical thickness, and $z'_{np}$ is the neutral plane depth. 

This notebook evaluates this relationship based on the results of a numerical model. Because the numerical model provides access to the vertical normal stress $\sigma_{zz}$, we can calculate the trench pull force  directly. 

---------------------

**Some further points**

- It is the  only location of the columns in Equation **(1)** & **(2)** that differentiates the trench pull force, from e.g, ridge push; both are net horizontal force components that arise due to differences described by the quantity $\Delta \bar\sigma_{zz}$.

- In an idealised 2D subducting plate, these 2 contributions (ridge push and trench pull) combine in a non-overlapping way: the ridge push captures the net force due to the isostatic topography, while the trench pull captures the net force  due to the non-isostatic topography.

- the absolute value of $F_{D}$ has meaning: it defines (in an overall sense) whether the stress state is tending to extend $F_{D}>0$ or shorten $F_{D}<0$ the lithosphere. Because significant differential stress can be supported only where $z' < z'_m$, the value of $F_{D}$ (i.e., the integral) should converge to a constant value (approximately independent of $z_c$).

- the magnitude of $\mathrm{GPE}^*$ in a column has no direct significance in terms of the horizontal force balance - only differences between columns matter. Unlike $F_{D}$, magnitude of $\mathrm{GPE}^* = \bar\sigma_{zz}$ in a single column always increases if we choose a larger integration depth. However, the integral of the $\Delta \mathrm{GPE}^* = \Delta \bar\sigma_{zz}$ (difference between columns) should converge with integration depth - if the assumption about the negligible pressure (or vertical normal stress) gradients at the compensation level is valid.

- In the MDOODZ7.0 code, the convention is that $z$ is positive up. In this notebook, the $z$ coordinate is switched for consistency with the manuscript. In terms of the signs of the stress tensor components, the only other required change in sign is for the shear stress.

- the origin of the z coordinate in this analysis is the initial level of the rock-sticky-air interface, the upper limit of the vertical integrations is the top of the model domain ($z=$-20 km), inclusive of the sticky-air layer. The sticky air has zero density, so $(\rho_m - \rho_w) = \rho_m =$ 3250 $\text{kg/m}^{3}$.

# Part 2: Flexural support of trench topography and $\Delta \mathrm{GPE}^*$ estimate

In this section we:

* estimate the "flexurally supported" topography
* calculate the location of the nearest isostatic column to the trench
* estimate the trench pull force based on the scaling relationship in the manuscript. 

In the classic thin plate flexure model, the vertical force balance is typically expressed as:

$\frac{dV}{dx} = -q$

where $V$ is the vertical shear stress resultant, and the normal load (*q*) is the given by the isostatic restoring force: $q = (\rho_m - \rho_w) g w(x)$. 

Here, $V$ is approximated by integrating the vertical component of the traction $\tau_{zx}\cdot \hat n$ down to the assumed compensation level $z_c$, and the horizontal gradient ($\frac{dV}{dx}$) is approximated using the numpy gradient function. We can then compare the "observed" model topography variation in the plate seaward of the trench, with the topographic variation that is implied by the gradients in $V$. 

The condition ($\frac{dV}{dx} = 0$) implies that the column experiences no flexural support. Seaward of the trench, the first location at which this condition met is often called the "first zero crossing". This location will be symbolised with $x_{I}$. 



```python
#integrate shear stresses along z axis using trapezoidal rule 
#neg 1 accounts for the unit vector
Vx = -1*np.trapz(tau_xz_grid[vint_mask, 1:], dx=1e3, axis =0)
#gradient along z usign central difference
dVdx = np.gradient(Vx)
#calc the no-isostatic delection (w) supported by the estimate of dVdx
vtopo = dVdx/(delrho*g)
```

```python
#get the first zero crossing. Closest index appoach is sufficient here.
#use a reasonable window (meaning here, kms) to search for the crossing
window=500
indx = np.argmin(-1*np.sum(tau_xz_grid, axis=0)[vxpts > tloc][0:window])
xi_loc = vxpts[vxpts > tloc][0:window][indx]
xi_indx = np.argmin(np.abs(vxpts - xi_loc)) 
xi_dist = xi_loc - tloc
```

```python
#calculate the topographic difference between the first zero crossing and the trench,
#note that this is not the total topography between the trench and outer rise,
#in the paper this this is $w_T$

trench_relative_topo = topo[:,1][np.argmin(abs(topo[:,0] - xi_loc))] - topo[:,1][np.argmin(abs(topo[:,0] - tloc))]

#get the index of the mid point between trench and first zero crossing 
xm_indx = int(tindx +   0.5*(xi_indx- tindx))
```

```python
#this is a depth mask to limit the search for the neutral plane, 
#or the zero-crossing of the function tau_xx - tau_zz 
np_mask = np.logical_and(czpts>10e3, czpts<50e3)

# Extract the masked array and corresponding x-values
z_vals = diff_stress_grid[:, xm_indx][np_mask]
x_vals = czpts[np_mask]

# Find indices where the sign changes (zero crossing)
sign_changes = np.where(np.diff(np.sign(z_vals)))[0]

# If a zero crossing exists, interpolate
if len(sign_changes) > 0:
    idx = sign_changes[0]  # Take the first crossing
    x1, x2 = x_vals[idx], x_vals[idx + 1]
    y1, y2 = z_vals[idx], z_vals[idx + 1]

    # Linear interpolation for zero crossing
    np_depth = x1 - y1 * (x2 - x1) / (y2 - y1)
    print(f"Interpolated neutral plane crossing at z = {1e-3*np_depth:.1f} km")
else:
    print("No zero crossing found within the specified range.")
```

```python
#estimate of the `trench pull force' using the scaling relationship from the paper
gpe_est = trench_relative_topo*delrho*g*np_depth

print(f"Trench Relative Topo: {1e-3*trench_relative_topo:.2f} km")
print(f"Neutral plane depth: {1e-3*np_depth:.2f} km")
print(f"Trench pull force (scaling estimate): {gpe_est*1e-12:.2f} TN/m")
```

## Part 3: Horizontal force balance

We will now plot the variation of the terms that appear in the vertically integrated, horizontal force balance, i.e: 


\begin{equation}
\Delta F_{D}  -  \Delta \mathrm{GPE}^*  +  F_B = 0 \;\;\mathrm{{\bf(3)}}
\end{equation}

In the manuscript, the equation is derived for 2 arbitrary points ($x_0$, $x_1$), and the equation simply says that 3 quantities must cancel (static equilibrium, or horizontal force balance). 

If we fix $x_0$, and allow $x_1$ to be a variable, then we can regard these 3 quantities a functions of $x$.  We define $\Delta$  as an operator, such that: $\Delta(f,x_0) = f(x) - f(x_0)$. 
We can now expand $\Delta F_{D}$ , and rewrite (3) as:

\begin{equation}
F_{D}(x)  -  \left( \Delta \mathrm{GPE}(x)^* -  F_{D}(x_0)  \right) +  F_B(x) = 0 \;\;\mathrm{{\bf(4)}}
\end{equation}

In other words, we will plot actual the $F_{D}(x)$, and use term ($-  F_{D}(x_0))$ to shift the value of $ \Delta \mathrm{GPE}(x)$. 

We do this because $F_D(x)$ has a meaningful value for any given point x: It represents the tendency of the integrated in-plane differential stress to extend ($F_D> 0$) or shorten the lithosphere ($F_D<0$). If we plotted the variation in terms as described by (3) all terms would be zero at $x_0$ and we loose the information about the actual value of $F_{D}(x)$. 

The basal term represents a cumulative integral, and is necessarily zero at $x_0$. 

In the plot figure below, $x_0=x_T$ (the trench axis).

All integrals are approximated using trapezoidal integration.



```python
#calculate the integrated qunatities using Trapezoidal function


#basal shear term
#set to zero at the trench
FB_x = 1e-12*((cumtrapz(tau_xz_grid[vbase_indx,:], x=vxpts))[:] \
        -(cumtrapz(tau_xz_grid[vbase_indx,:], x=vxpts))[:][tindx])


#in-plane differential stress term
#instead, the GPE^* is referenced to F_D at the trench 
del_FD_x = 1e-12*np.trapz(diff_stress_grid[cint_mask, :], x=czpts[cint_mask], axis=0) 



#note this $\Delta {GPE^*} = - \bar \sig_{zz}$, not the true GPE $\bar P_L$, which has no relavance in non-isostatic settings
#the first 2 lines calculate the difference in vertically integrated vertical norma stress
#the final line normalised this to the same valeu as F_D at the trench
del_GPE_x =-1*( 1e-12*np.trapz(sig_zz_grid[cint_mask, :], x=czpts[cint_mask], axis=0) - 
          1e-12*np.trapz(sig_zz_grid[cint_mask, :], x=czpts[cint_mask], axis=0)[tindx] 
           - del_FD_x[tindx])


# \Delta \bar \sig_{zz}$
#this is the variation on the `full' horizontal normal stress resultant, 
#which is equal to \Delta F_D - \Delta {GPE^*} 
del_bar_sig_xx =  -1*( 1e-12*np.trapz(sig_xx_grid[cint_mask, :], x=czpts[cint_mask], axis=0) - 
                  1e-12*np.trapz(sig_xx_grid[cint_mask, :], x=czpts[cint_mask], axis=0)[tindx])
```

**comments**

For the reference timestep, the quantity $F_D$ is near-neutral at the trench, and becomes increasingly positive between the trench and the outer rise, where it peaks. Between the trench and the outer rise, the increase in $\Delta F_{D}$ corresponds to increasing (deviatoric) tension and a net force to the right.

On the length scale over which the trench topography is developed, ($\sim$ 50-100 km), we have $\Delta F_{D} \approx \Delta \mathrm{GPE}^*$ which simply reflects the fact the integrated basal drag ($F_B$) is negligible over this length scale.

The presence of a (deviatoric) tension $\Delta F_{D}  > 0$ in subducting plates does not necessarily imply that this is tension being transmittted "through the hinge". For the reference timestep shown here, the (deviatoric) tension in the plate seaward of the trench is associated with the trench pull $\Delta  \mathrm{GPE}^*$. 

At different timesteps in this model, the value of $F_{D}$ varies quite significantly, and at many stages reflects a deviatoric compression $\Delta F_{D}  < 0$. However you will still see the change ($\Delta F_{D}$) that is associated with the trench pull force $\Delta \mathrm{GPE}^*$. 



## Part 4 : Depth variation of stress

The following plots show the depth distribution of various quantities of interest at key points in the trench - outer slope region. 

The most important panel is **c** showing the difference in the vertical normal stress between $x_{I}$ and the trench ($x_T$). That is: 

\begin{equation}
\Delta \sigma_{zz}(z) = \sigma_{zz}(x_{I}, z) - \sigma_{zz}(x_{T}, z)
\end{equation}

When the manuscript talks about the "equilibration of vertical normal stress" it is referring to the fact that this difference diminishes with depth. The trench pull force is equal to the area bounded by $\Delta \sigma_{zz}(z)$.

In the figure below, a smoothing operator is used to address high-frequency noise caused by stress rotations in shear bands, which result from pressure-dependent plasticity. In the numerical model, small stress rotations create noise in the vertical shear stress pattern, which is reflected in the vertical normal stress (which is sensitive to the horizontal gradient in vertical shear stress). 

 The shear bands create noise that oscillates around zero, having negligible impact on the vertically integrated value. The smoothing window, set by `smoothwin` averages successive grid cells either side of the (cells are spaced 1 km apart) to help to suppress this noise.  Note that the magnitude of the trench pull force is fundamentally mediated by the depth of equilibration of $\sigma_{zz}(z)$. This is sensitive to integrated values of the (horizontal) gradients of the vertical shear stress. The noise created by the shear bands largely cancels out. 


```python
smoothwin = 10

#get the vertical normal stress around the isostatic column loc. 
sig_zz_iso = np.mean(sig_zz_grid[:, xi_indx - smoothwin : xi_indx + smoothwin], axis=1) if smoothwin else sig_zz_grid[:, xi_indx]

#get the veertical normal stress at the trench column  loc. 
sig_zz_trench = np.mean(sig_zz_grid[:, tindx - smoothwin : tindx + smoothwin], axis=1) if smoothwin else sig_zz_grid[:, tindx]



#get vertical profiles of the temp, in-plane differential stress and vertical shear stress
#these are taken at the mid-point between the trench and the isostatic loc.
temp_profile = np.mean(temp_grid[:, xm_indx - smoothwin:xm_indx + smoothwin], axis=1) if smoothwin else temp_grid[:, xm_indx]
diff_stress_profile = np.mean(diff_stress_grid[:, xm_indx - smoothwin : xm_indx + smoothwin], axis=1) if smoothwin else diff_stress_grid[:, xm_indx]
vertical_shear_stress_profile = np.mean(tau_xz_grid[:, xm_indx - smoothwin : xm_indx + smoothwin], axis=1) if smoothwin else tau_xz_grid[:, xm_indx]
```

```python
def pv_plane_from_h5(file, xleft=0, zbase =0):
    
    data   = file['/Model/Params']
    nx             = int(data[3])                   #/ No. grid points x
    nz             = int(data[4]) 
    
    xspan = data[1]
    zspan = data[2]
    
    center = (xspan/2. +  xleft, 
              zbase - zspan/2.,
              0)

    plane = pv.Plane(
        center=center,
        direction=(0, 0, -1),
        i_size=xspan,
        j_size=zspan,
        i_resolution=nx-1,
        j_resolution=nz-1,
    )
    
    return plane

def lines_from_points(points):
    """Given an array of points, make a line set"""
    poly = pv.PolyData()
    poly.points = points
    cells = np.full((len(points) - 1, 3), 2, dtype=np.int_)
    cells[:, 1] = np.arange(0, len(points) - 1, dtype=np.int_)
    cells[:, 2] = np.arange(1, len(points), dtype=np.int_)
    poly.lines = cells
    return poly
```
